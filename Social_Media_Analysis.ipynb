## 1. Age vs Social Media Consumption
* How does social media usage vary across different age groups?

```python
# Deciding bins size
bins = [10, 20, 30, 40, 50, 100]
# Deciding bin labels
labels = ['11-20', '21-30', '31-40', '41-50','60+']

# Grouping data according to Bins and labels
df['age_group'] = pd.cut(df['age'],bins=bins,labels=labels)

# Create a figure with a custom size
fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # Width=12 to fit two 6-inch plots

# Creating a boxplot of the x-axis =  'age' and y-axis = 'total_time_spent' columns of the dataframe
# --- Plot 1: box Plot ---
# Direct to the first subplot
sns.boxplot(x='age_group',y='total_time_spent',data=df,showmeans=True,ax=axes[0] )

# Add a title to the plot
axes[0].set_title("Time Spent by Age Group")
# Add labels to the x-axis and y-axis
axes[0].set_xlabel("Age Group")
axes[0].set_ylabel("Average Time")

# --- Plot 2: Line Plot ---
age_mean = df.groupby('age_group')['total_time_spent'].mean()
age_mean.plot(kind='line', marker='o', ax=axes[1]) # Direct to the second subplot

# Add a title to the plot
axes[1].set_title("Average Time Spent by Age Group")
# Add labels to the x-axis and y-axis
axes[1].set_xlabel("Age Group")
axes[1].set_ylabel("Average Time")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**The boxplot shows that users aged 11â€“20 have the highest median and mean time spent, indicating the strongest engagement among teenagers. Usage declines steadily from 21â€“50, where both median and average time spent are lowest. Interestingly, users aged 60+ show a noticeable increase in average usage, suggesting renewed engagement among senior citizens. The wide spread and presence of outliers in younger groups indicate more variability in their usage patterns. Overall, social media consumption follows a U-shaped trend, being highest among teenagers and seniors, and lowest among middle-aged users.**

```python
plt.figure(figsize = (6,5))
# Deciding bins size
bins = [12, 22, 32, 42, 52, 100]
# Deciding bin labels
labels = ['11-22', '22-32', '32-42', '42-52','60+']

# Grouping data according to Bins and labels
df['age_group'] = pd.cut(df['age'],bins=bins,labels=labels)
ax = sns.countplot(x='age_group', data=df, stat= 'percent')
for c in ax.containers:
    ax.bar_label(c,fmt='%.1f%%')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Users')

plt.title('Distribution of user by Age Group')

plt.show()
```

**The largest user group falls in the 11â€“22 age range (24%), followed by 32â€“42 (20.5%) and 22â€“32 (20%), indicating strong participation from young and middle-aged users. Participation gradually declines with age, with the 60+ group being the smallest (16.1%).**

## 5. Addiction Level Distribution
* What proportion of users fall into low, medium, and high addiction categories?



```python
def addiction_group(x):
    if x <= 3:
        return "Low"
    elif x <= 6:
        return "Medium"
    else:
        return "High"

df['addiction_category'] = df['addiction_level'].apply(addiction_group)
```

```python
plt.figure(figsize = (12,5))
sns.countplot(x = 'addiction_category',data = df, hue = 'addiction_category',legend = False, palette='Set2')

plt.ylabel('Total Users')
plt.xlabel('Addiction_level')

plt.title('Distribution of Addiction levels of Users')

plt.show()
```

```python
df['addiction_category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

**The distribution shows that most users fall into the high addiction category, followed by medium, while a small proportion belongs to the low category. This indicates that a majority of users exhibit strong dependence on social media platforms.**
